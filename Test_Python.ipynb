```python
# Indetify the What is the mean of ‘average_rating’ for the books where ‘language_code’ is ‘eng’ using groupby
books_df = pd.read_csv("/content/books.csv")
print(books_df[books_df["language_code"] == "eng"]["average_rating"].mean())

```
